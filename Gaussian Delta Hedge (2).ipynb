```python
import numpy as np
import scipy as scipy
import pandas as dp
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
def generate_path(start, sigma, num_divisions):
    path = [start]
    for i in xrange(num_divisions):
        path.append(path[-1] + sigma*np.random.normal(0, np.sqrt(1.0/num_divisions)))
    return path
```

```python
path = generate_path(0.0, 1.0, 1000)
plt.plot(path)
plt.show()
plt.plot([path[k]-path[k-1] for k in xrange(1,len(path))])
plt.show()
```

```python
def straddle_delta(time_to_exp, sigma, price, strike):
#     d1 = (np.log(price/strike) + (sigma**2)*time_to_exp/2.0)/(sigma*np.sqrt(time_to_exp))
    d1 = (price - strike)/(sigma*np.sqrt(time_to_exp))
    return 2*norm.cdf(d1) - 1
```

```python
x = np.arange(95.0,105,0.01)
plt.plot(x,[straddle_delta(1.0, 0.5, y, 100.0) for y in x])
plt.show()
```

```python
def generate_scenario(start, sigma, num_divisions):
    t = 0
    path = [start]
    shares = -straddle_delta(1.0, sigma, start, start)
    
    call = (sigma/np.sqrt(2*np.pi))
    
    money = -shares*path[-1] - (2*call)
    money_path = [money]
    shares_path = [shares]
    changes=[]
    max_change = 0
    for i in xrange(num_divisions-1):
        path.append(path[-1] + sigma*np.random.normal(0, np.sqrt(1.0/num_divisions)))
        if abs(path[-1]-path[-2])>abs(max_change):
            max_change = path[-1]-path[-2]
        changes.append(path[-1]-path[-2])
        t = t + 1.0/num_divisions
        hedge_shares = -(straddle_delta(1-t, sigma, path[-1], start))
#         print path[-1], 1-t, hedge_shares
        money = money - (hedge_shares - shares)*path[-1]
        money_path.append(money)
        shares = hedge_shares
        shares_path.append(shares)
    path.append(path[-1] + sigma*np.random.normal(0, np.sqrt(1.0/num_divisions)))
    money = money + shares*path[-1]
    shares = 0
    money = money + abs(start - path[-1])
    money_path.append(money)
    return money, path, money_path, shares_path, max_change, np.mean(changes), np.mean(map(lambda a:abs(a),changes)), sum(changes)
```

```python
money, path, money_path, shares_path, max_change, avg_change, avg_abs_change, sum_changes = generate_scenario(100, 0.5, 2000)
plt.plot(path)
plt.show()
plt.plot(shares_path)
plt.show()
print money, path[-1]-path[0]
#print money_path
```

```python
def simple_scenario(start, sigma, num_divisions):
    t = 0
    path = [start]
    shares = -straddle_delta(1.0, sigma, start, start)
    
    call = (sigma/np.sqrt(2*np.pi))
    
    money = -shares*path[-1] - (2*call)
    money_path = [money]
    shares_path = [shares]
    for i in xrange(num_divisions-1):
        path.append(path[-1] + sigma*np.random.normal(0, np.sqrt(1.0/num_divisions)))
        t = t + 1.0/num_divisions
        hedge_shares = -(straddle_delta(1-t, sigma, path[-1], start))
#         print path[-1], 1-t, hedge_shares
        money = money - (hedge_shares - shares)*path[-1]
        money_path.append(money)
        shares = hedge_shares
        shares_path.append(shares)
    path.append(path[-1] + sigma*np.random.normal(0, np.sqrt(1.0/num_divisions)))
    money = money + shares*path[-1]
    shares = 0
    money = money + abs(start - path[-1])
    money_path.append(money)
    return money
```

```python
def generate_sabr(start, start_vol, alpha, beta, rho, num_divisions):
    path = [start]
    path_vol = [start_vol]
    for i in xrange(num_divisions):
        dW1 = np.random.normal(0, np.sqrt(1.0/num_divisions))
        dW2 = rho*dW1 + np.sqrt(1-rho**2)*np.random.normal(0, np.sqrt(1.0/num_divisions))
        path.append(path[-1] + path_vol[-1]*(path[-1]**beta)*dW1)
        path_vol.append(path_vol[-1] + alpha*path_vol[-1]*dW2)
    return path, path_vol
```

```python
path, path_vol = generate_sabr(100, 1.0, 1.0, 1.0, 0.5, 10000)
```
